```python
#mlxtend 라이브러리에서 apriori, association_rules 기능을 불러옴
from mlxtend.frequent_patterns import apriori, association_rules

#apriori 알고리즘을 통해 Frequent Itemsets 계산
#오렌지3 Frequent Itemsets 환경 설정과 동일하게 설정
frequent_itemsets = apriori(df, min_support = .01, max_len = 6, use_colnames = True)

# Frequent Itemset를 support값에 따라 정렬한 후 첫 5행 출력
frequent_itemsets.sort_values(by=['support']).head()
```

```python
#Frequent_Itemsets를 대상으로 연관 규칙(association rule) 생성
rules = association_rules(frequent_itemsets, metric = 'support', min_threshold = 0.01)

c1 = rules['confidence'] > 0.85 #조건 C1 생성
filtered_rules = rules[c1] #조건 C1을 만족하는 연관 규칙 필터링

#support 변수를 기준으로 내림차순 정렬
filtered_rules.sort_values(by=['support'], ascending = False)
```
